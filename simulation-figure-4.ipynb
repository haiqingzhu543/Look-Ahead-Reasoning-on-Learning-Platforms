```python
# === Simulation: Logistic regression (no intercept)
# === Strategic D1 flip+move toward _D2
# === Utility u_i = -|| - _target||

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from data_prep import load_data
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
import matplotlib

def sigmoid(z):
    return 1 / (1 + np.exp(-z))



# ----- Data generation -----
X, Y, data = load_data('./cs-training.csv')
X = X[:, 1:]

strat_ind = [1, 5]

# distributional outliers:
X, Y = X[X[:,strat_ind[0]] < 10], Y[X[:,strat_ind[0]] < 10]
X, Y = X[X[:,strat_ind[1]] < 10], Y[X[:,strat_ind[1]] < 10]

n = X.shape[0]
d = X.shape[1]

idx_D1, idx_D2= np.where(Y == 1)[0], np.where(Y == 0)[0]
n1, n2 = len(idx_D1), len(idx_D2)
X1, X2 = X[idx_D1, :].reshape((n1, d)), X[idx_D2, :].reshape((n2, d))
mu_D1, mu_D2 = np.mean(X1, axis = 0), np.mean(X2, axis = 0)

clf0 = LogisticRegression(fit_intercept=False, solver="lbfgs", max_iter=1000)
clf0.fit(X, Y)
theta0 = clf0.coef_[0]



# ----- Parameters -----
eta = 0.552
alphas = np.linspace(0.1, 0.99, 20)
n_trials = 10
X_all = X
y_all = Y


# ----- Utility -----
def compute_utilities(theta, theta_target):
    return -np.linalg.norm(theta - theta_target)**2

# ----- Strategy -----
def apply_strategy(X, idx_h, mu_ref, eta):
    X_new = X.copy()
    mask = np.zeros(X.shape[1])
    mask[strat_ind] = 1

    # Only modify rows in D (idx_D1) that are selected (idx_h)
    X_new[idx_h] = X[idx_h] + eta * (mu_ref * mask)
    return X_new


# ----- Gradients / Hessian -----
def compute_gradients_and_hessian(X_all, y_all, theta, X_sub, y_sub):
    """
    Logistic regression (no intercept):
      _i = -[y log (x) + (1-y) log (1-(x))]
       = ((x) - y)x
      H = E[(1-)xx]
    """
    if len(X_sub) > 0:
        p_sub = sigmoid(X_sub @ theta)
        grad_u = -(theta - theta_target)
        grad_l = -((p_sub - y_sub)[:, None] * X_sub).mean(axis=0) + theta
    else:
        grad_u = np.zeros(X_all.shape[1])
        grad_l = np.zeros(X_all.shape[1])
    p_all = sigmoid(X_all @ theta)
    w = p_all * (1 - p_all)
    H = (X_all * w[:, None]).T @ X_all / X_all.shape[0]
    H = H + np.eye(H.shape[0])
    return grad_u, grad_l, H


alpha_targets = [0.3, 0.8]
theta_targets = []

for alpha in alpha_targets:
    k = int(round(alpha * n1))
    idx_flip = np.random.choice(idx_D1, size=k, replace=False)
    X_new = apply_strategy(X, idx_flip, mu_D2, eta)
    y_new = Y

    # logistic regression without intercept
    clf = LogisticRegression(fit_intercept=False, solver="lbfgs", max_iter=1000, C= 1, tol = 1e-6)
    clf.fit(X_new, y_new)
    theta = clf.coef_[0]
    theta_targets.append(theta)

theta_target = theta_targets[0]
etas = [0.559, 0.2159, 0.3400]

# ----- Main sweeps -----
rows = []
for eta in etas:
    for alpha in alphas:
        vals_inner, vals_util = [], []
        for i in range(n_trials):
            # select  fraction of D1
            rng = np.random.default_rng(i)
            k = int(round(alpha * n1))
            idx_flip = rng.choice(idx_D1, size=k, replace=False)
            X_new = apply_strategy(X, idx_flip, mu_D2, eta)
            y_new = Y

            # logistic regression without intercept
            clf = LogisticRegression(fit_intercept=False, solver="lbfgs", max_iter=1000, C= 1, tol = 1e-6)
            clf.fit(X_new, y_new)
            theta = clf.coef_[0]

            # gradients
            X_sub, y_sub = X_new[idx_flip], y_new[idx_flip]
            gu, gl, H = compute_gradients_and_hessian(X_new, y_new, theta, X_sub, y_sub)
            try:
                v = np.linalg.solve(H, gl)
            except np.linalg.LinAlgError:
                v = np.linalg.solve(H + 1e-6*np.eye(H.shape[0]), gl)
            vals_inner.append(float(gu @ v))
            
            # new utility based on  distance to _target
            u_val = compute_utilities(theta, theta_target)
            vals_util.append(u_val)
        
        rows.append({
            "eta": eta,
            "alpha": alpha,
            "mean_inner": np.mean(vals_inner),
            "std_inner": np.std(vals_inner, ddof=1),
            "mean_util": np.mean(vals_util),
            "std_util": np.std(vals_util, ddof=1),
            "is_negative": np.mean(vals_inner) < 0
        })

df = pd.DataFrame(rows)
print(theta_target)
# Plotting fonts setup
matplotlib.rcParams['mathtext.fontset'] = 'cm'
matplotlib.rcParams['font.family'] = 'Times New Roman'
matplotlib.rcParams.update({'font.size': 16}) 

# ----- Plot 1: Alignment metric -----
plt.figure()
df1 = df.query("eta == 0.559")
plt.errorbar(df1.alpha, df1.mean_inner, yerr=df1.std_inner, fmt='-o', capsize=3, label = r"$h^\sharp_{\alpha = 0.3}$")
df3 = df.query("eta == 0.3400")
plt.errorbar(df3.alpha, df3.mean_inner, yerr=df3.std_inner, fmt='-*', capsize=3, label = r"$h^\sharp_{\alpha = 0.5}$")
df2 = df.query("eta == 0.2159")
plt.errorbar(df2.alpha, df2.mean_inner, yerr=df2.std_inner, fmt='-x', capsize=3, label = r"$h^\sharp_{\alpha = 0.8}$")

plt.axhline(0, ls='--', color='gray')
plt.xlabel("")
plt.ylabel(r"Alignment Metric $\Psi$")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("alignment_metric_fix_strat.pdf")
plt.savefig("alignment_metric_fix_strat.png")
plt.show()

# ----- Plot 2: Utility vs  -----
plt.figure()

plt.errorbar(df1.alpha, df1.mean_util, yerr=df1.std_util, fmt='-o', capsize=3, label = r"$h^\sharp_{\alpha = 0.3}$")
plt.errorbar(df3.alpha, df3.mean_util, yerr=df3.std_util, fmt='-*', capsize=3, label = r"$h^\sharp_{\alpha = 0.5}$")
plt.errorbar(df2.alpha, df2.mean_util, yerr=df2.std_util, fmt='-x', capsize=3, label = r"$h^\sharp_{\alpha = 0.8}$")

plt.xlabel("")
plt.ylabel(r"Utility $U$")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("utility_fix_strat.pdf")
plt.savefig("utility_fix_strat.png")
plt.show()

# ----- Plot 3: Decision boundaries -----
selected_alphas = [0.1, 0.4, 0.7, 0.99]
fig, axes = plt.subplots(1, len(selected_alphas), figsize=(20, 4))
xid, yid = strat_ind  # indices of strategic features

# feature ranges
x_min, x_max = X[:, xid].min() - 0.5, X[:, xid].max() + 0.5
y_min, y_max = X[:, yid].min() - 0.5, X[:, yid].max() + 0.5
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 200),
    np.linspace(y_min, y_max, 200)
)

for i, alpha in enumerate(selected_alphas):
    k = int(round(alpha * n1))
    idx_flip = rng.choice(idx_D1, size=k, replace=False)
    X_new = apply_strategy(X, idx_flip, mu_D2, eta)
    y_new = Y

    clf = LogisticRegression(fit_intercept=False, solver="lbfgs", max_iter=1000)
    clf.fit(X_new, y_new)
    theta = clf.coef_[0]

    print(theta)

    # --- decision function on the 2D grid ---
    # project only on selected features, keep others fixed at their means
    X_ref = np.zeros((xx.size, d))
    X_ref[:, :] = X.mean(axis=0)
    X_ref[:, xid] = xx.ravel()
    X_ref[:, yid] = yy.ravel()
    zz = sigmoid(X_ref @ theta).reshape(xx.shape)

    ax = axes[i]
    # contour decision boundary ( = 0.5)
    cs = ax.contour(
        xx, yy, zz, levels=[0.5], colors='k', linewidths=2, linestyles='--'
    )

    # scatter D and D (projected on strat features)
    ax.scatter(X_new[idx_D1, xid], X_new[idx_D1, yid], s=8, c='blue', alpha=0.3, label='D1')
    ax.scatter(X_new[idx_D2, xid], X_new[idx_D2, yid], s=8, c='red',  alpha=0.3, label='D2')

    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xlabel(f"Feature {xid}")
    ax.set_ylabel(f"Feature {yid}")
    ax.set_title(f" = {alpha:.2f}")
    ax.grid(True)
axes[0].legend(loc='upper right')
fig.suptitle(
    f"Decision boundaries on strategic features [{xid}, {yid}] "
    f"(={eta}, Logistic Regression no intercept)\n"
    r"$u_i = -\|\theta - \theta_{target}\|^2$",
    fontsize=13
)
plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()

df
```
